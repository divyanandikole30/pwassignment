```python
#remove or replace null values from  job_type
df['job_type'].value_counts()
```

```python
cat_col=df.select_dtypes(include='object').columns
num_col=df.select_dtypes(exclude='object').columns
```

```python
#convert categorical columns to numerical columns
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in cat_col:
  df[col]=le.fit_transform(df[col])
```

```python
#perform data encoding on this on type_of_city
df['type_of_city'].value_counts()

from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
encoded_data=ohe.fit_transform(df[['type_of_city']])
encoded_df=pd.DataFrame(encoded_data.toarray(),columns=ohe.get_feature_names_out(['type_of_city']))
df=pd.concat([df,encoded_df],axis=1)
```
