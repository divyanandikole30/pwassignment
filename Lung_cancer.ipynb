```python
#perform OneHot Encoding on Gender column
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
gender_encoded=ohe.fit_transform(df[['GENDER']])
gender_encoded_df=pd.DataFrame(gender_encoded.toarray(),columns=ohe.get_feature_names_out(['GENDER']))
df=pd.concat([df,gender_encoded_df],axis=1)
df.drop('GENDER',axis=1,inplace=True)
```

```python
df.info()
```

```python
#Convert 'Lung_cancer ' columns with Yes and no values into numerical column
df['LUNG_CANCER']=df['LUNG_CANCER'].map({'YES':1,'NO':0})
```
